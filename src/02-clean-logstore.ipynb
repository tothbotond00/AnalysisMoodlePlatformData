# Clean the logstore
Duplicates exist in the logstore. They are removed, as a most basic clean up.

```python
import pandas as pd

df = pd.read_csv("./log/logstore_standard_log_anonymized.csv")
```

```python
display(df.describe())
df_cleaned = df.drop_duplicates()
display(df_cleaned.describe())
```

```python
df_cleaned.to_csv("./log/logstore_standard_log_anonymized_no_duplicates.csv", index=False)
```
