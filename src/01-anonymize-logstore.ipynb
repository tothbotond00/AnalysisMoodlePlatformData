# Anonymize the logstore CSV
Moodle provided data from a course, the original data that Moodle provided was not completely anonymized. Therefore this notebook has been created for anonymization. This can not be reproduced, since the original logstore file has been deleted after creating an anonymous one, for privacy protection.

```python
import pandas as pd

df = pd.read_csv("./log/mdl_logstore_standard_log.csv") # the original logstore, now deleted.
```

In the original dataset, "other" contains meta data, e.g. forum posts, where ppl introduce themselves with full name, job and city. So we replace "other" with default empty object "{}" or keep "N;".

```python
df["other_m"] = df["other"]=="N;"
df["other_m"] = df["other_m"].replace({True: "N;", False: "{}"})
df["other"] = df["other_m"]
df.drop("other_m", axis=1, inplace=True)
df
```

```python
df.to_csv("log/logstore_standard_log_anonymized.csv", index=False)
```
